```python
import requests
import urllib.request
def kakao_search_place(addr):
    api_key = 'KakaoAK 5bc7785bedf372180f58929a0899232f'
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?&query='
    query = urllib.parse.quote(addr)
    result = requests.get(url+query, headers={'Authorization': api_key}).json()
    #print(result)
    results =[]
    if len(result['documents']) != 0:
        for i in range(len(result['documents'])):
            if result['documents'][i]["address_name"] !="":
                results.append(result['documents'][i]["place_name"])
                results.append(result['documents'][i]["category_name"])
                results.append(result['documents'][i]["address_name"])
                results.append(result['documents'][i]["x"])
                results.append(result['documents'][i]["y"])
    else:
        results.append("no results")
    return results
```
