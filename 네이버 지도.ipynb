```python
import urllib.request
import json

def naver_reverse_geocode(lat, lon):

    #애플리케이션 클라이언트 id 및 secret
    client_id = "cgpixbclor"
    client_secret = "MITMhvEHsoZJ2SZhN00kWXtaE1aKbLMRB8QRG9LD"

    url = "https://naveropenapi.apigw.ntruss.com/map-reversegeocode/v2/gc"
    option = "&sourcecrs=epsg:4326&orders=addr,roadaddr&output=json"
    location = str(lat) + "," + str(lon)
    query = "?request=coordsToaddr&coords="+urllib.parse.quote(location)

    url_query = url + query + option
    #Open API 검색 요청 개체 설정
    request = urllib.request.Request(url_query)
    request.add_header("X-NCP-APIGW-API-KEY-ID",client_id)
    request.add_header("X-NCP-APIGW-API-KEY",client_secret)

    #검색 요청 및 처리
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode == 200):
        response_body = response.read()
        response_result = response_body.decode('utf-8')    
        jsonObject = json.loads(response_result)
        if len(jsonObject["results"]) == 1:
            print("지번 주소 -> " + get_addr(jsonObject))
        else: 
            print("지번 주소 -> " + get_addr(jsonObject))
            print("도로명 주소 -> " + get_roadaddr(jsonObject))
    else:
        print("Error code:"+rescode)
```

# Json 파일 파싱 -> 주소 추출

```python
def get_addr(jsonObject):
    addr_list = []
    for j in jsonObject.get("results")[0].get("region").keys():
        if(jsonObject["results"][0]["region"][j]["name"] != ""):
            addr_list.append(jsonObject["results"][0]["region"][j]["name"])
    if jsonObject["results"][0]["land"]["number1"] != "" and jsonObject["results"][0]["land"]["number2"] != "":
        addr_list.append(jsonObject["results"][0]["land"]["number1"] + "-" + jsonObject["results"][0]["land"]["number2"])
    elif jsonObject["results"][0]["land"]["number1"] != "" and jsonObject["results"][0]["land"]["number2"] == "":
        addr_list.append(jsonObject["results"][0]["land"]["number1"])
    addr = ' '.join(addr_list[1:])
    return addr
```

```python
def get_roadaddr(jsonObject):
    roadaddr_list = []
    for j in jsonObject.get("results")[1].get("region").keys():
        if(jsonObject["results"][1]["region"][j]["name"] != ""):
            roadaddr_list.append(jsonObject["results"][1]["region"][j]["name"])
    roadaddr_list.append(jsonObject["results"][1]["land"]["name"])
    roadaddr_list.append(jsonObject["results"][1]["land"]["number1"])
    if jsonObject["results"][1]["land"]["addition0"]["value"] != "":
        roadaddr_list.append(jsonObject["results"][1]["land"]["addition0"]["value"])
    roadaddr = ' '.join(roadaddr_list[1:])
    return roadaddr
```
