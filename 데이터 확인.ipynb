```python
def remove_0(df):
    for i in range(len(df)):
        if df['lat'][i] == 0 or df['lon'][i] == 0:
            df = df.drop(i)
        
    df = df.reset_index(drop=True)
    return df
```

```python
def get_xdata(df):
    x = df.iloc[:, [1, 2]].values
    return x
```
