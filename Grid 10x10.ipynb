```python
grid_size = 10
start_lat = 35.951225
start_lon = 126.677630592098

end_lat = 35.94053805
end_lon = 126.689017999998


result = (end_lon - start_lon) / grid_size
result1 = (start_lat - end_lat) / grid_size
```

```python
x = []
y = []
x.append(start_lon)
y.append(start_lat)
for i in range(grid_size):
    x.append(x[i] + result)
    y.append(y[i] - result1)
print(x)
print(y)
```
